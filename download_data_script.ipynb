```python
# Ran this to get buildings for each country (took 13.9 hours)
con = duckdb.connect()
con.install_extension("spatial")
con.load_extension("spatial")
con.sql("LOAD spatial;")
con.sql("SET s3_region='us-west-2';")

for target_country in countries_dict:

    country_name = str(countries_dict[target_country])
    con.sql(f"""
    CREATE OR REPLACE TABLE bounds AS (
        SELECT
            id AS division_id, names.primary, geometry, bbox
        FROM
            read_parquet('s3://overturemaps-us-west-2/release/2025-06-25.0/theme=divisions/type=division_area/*.parquet')
        WHERE
            subtype = 'country'
            AND country = '{target_country}'
    );
    """)

    con.sql("""
    SET variable xmin = (select bbox.xmin FROM bounds);
    SET variable ymin = (select bbox.ymin FROM bounds);
    SET variable xmax = (select bbox.xmax FROM bounds);
    SET variable ymax = (select bbox.ymax FROM bounds);
    SET variable boundary = (select geometry FROM bounds);
    """)

    con.sql(f"""
    COPY(
        SELECT
            id,
            names.primary as primary_name,
            height,
            subtype,
            class,
            geometry
        FROM
            read_parquet('s3://overturemaps-us-west-2/release/2025-06-25.0/theme=buildings/type=building/*.parquet')
        WHERE
            names.primary IS NOT NULL
            AND class IN (
                'commercial', 'factory', 'garage', 'garages', 'hotel', 'industrial',
                'kiosk', 'manufacture', 'office', 'parking', 'post_office',
                'retail', 'service', 'supermarket', 'warehouse'
            )
            AND height >=30
            AND bbox.xmin > getvariable('xmin')
            AND bbox.xmax < getvariable('xmax')
            AND bbox.ymin > getvariable('ymin')
            AND bbox.ymax < getvariable('ymax')
            AND ST_INTERSECTS(
                getvariable('boundary'),
                geometry
            )
    ) TO '{country_name}_filtered_buildings.geojson' WITH (FORMAT GDAL, DRIVER 'GeoJSON');
    """)
```

```python
# RUN IF YOU WANT TO DO IT BY CHUNKING (MAY TAKE AROUND 8 HOURS) (NO NEED TO DO THIS ANYMORE)
con = duckdb.connect()
con.install_extension("spatial")
con.load_extension("spatial")
con.sql("LOAD spatial;")
con.sql("SET s3_region='us-west-2';") # where s3 file is hosted, may be redundant

# Projection should be EPSG:4326 (WGS 84)
# xmin is longitude -180 to 180; half is 0 to 180
# ymin is latitude -90 to 90; half is 0 to 90

# NE hemisphere is 0 to 180 long and 0 to 90 lat (CURRENTLY SET ON NE HEMISPHERE)
# NW hemisphere is 0 to -180 long and 0 to 90 lat
# SE hemisphere is 0 to 180 long and 0 to -90 lat
# SW hemisphere is 0 to -180 long and 0 to -90 lat

# download in chunks of data 10 degrees by 10 degrees
# Faster to extract through azure
lon_list = []
for lon in range(0,180,10): # ADJUST HERE FOR DIFFERENT HEMISPHERE
    lon_list.append((lon, lon + 10))
lat_list = []
for lat in range(0,90,10):  # ADJUST HERE FOR DIFFERENT HEMISPHERE
    lat_list.append((lat, lat + 10))

for lat_min, lat_max in lat_list:
    for lon_min, lon_max in lon_list:
        
        outfile = f"buildings_{lon_min}_{lat_min}_{lon_max}_{lat_max}.geojson"
        print(f"Currently working on buildings_{lon_min}_{lat_min}_{lon_max}_{lat_max}...")
        con.sql(f"""
        COPY(
            SELECT
                id,
                names.primary as primary_name,
                height,
                subtype,
                class,
                geometry
            FROM
                read_parquet('az://overturemapswestus2.blob.core.windows.net/release/2025-06-25.0/theme=buildings/type=building/*', filename=true, hive_partitioning=1)
            WHERE
                names.primary IS NOT NULL
                AND class IN ('commercial', 'factory', 'garage', 'garages', 'hotel', 'industrial', 'kiosk', 'manufacture', 'office', 'parking', 'post_office', 'retail', 'service', 'supermarket', 'warehouse')
                AND height >= 30
                AND bbox.xmin BETWEEN {lon_min} AND {lon_max}
                AND bbox.ymin BETWEEN {lat_min} AND {lat_max}
        ) TO '{outfile}' WITH (FORMAT GDAL, DRIVER 'GeoJSON'); 
        """)
        print('Done!')

```
